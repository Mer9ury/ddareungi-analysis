```python
x_19=pd.merge(df19_7to10,df19_17to20, left_on='station', right_on='station', how='inner')
x_20=pd.merge(df20_7to10,df20_17to20, left_on='station', right_on='station', how='inner')
x_7to10=pd.merge(df19_7to10,df20_7to10, left_on='station', right_on='station', how='inner')
x_17to20=pd.merge(df19_17to20,df20_17to20, left_on='station', right_on='station', how='inner')
```

# 2019 출근-퇴근

```python
df_count=x_19.copy()[['station','보관소(대여소)명_x','자치구_x','Count_x','Count_y']]
df_count['Net']=df_count['Count_y']-df_count['Count_x']
df_count = df_count.sort_values(by='Net' ,ascending=False)
# 출근<<퇴근
df_count.head(10)
```

```python
df_count.tail(10)
```
