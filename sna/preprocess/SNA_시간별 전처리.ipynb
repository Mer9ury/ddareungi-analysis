```python
import pandas as pd
import datetime
```

```python
df = pd.read_csv('2020_time.csv')
```

```python
df
```

```python
df.drop(['return_time'],axis = 1, inplace = True)
```

```python
df_time = df['borrow_time']
df['borrow_date'] = df.borrow_time.str.split(' ').str[0]
df['borrow_time'] = df.borrow_time.str.split(' ').str[1]
```

```python
df['borrow_m'] = df.borrow_date.str.split('-').str[1].astype(int)
df['borrow_d'] = df.borrow_date.str.split('-').str[2].astype(int)

```

```python
df['borrow_time'] = df.borrow_time.str.split(':').str[0].astype(int)
```

```python
df_isweek = []
for i in range(df.shape[0]):
    df_isweek.append(datetime.date(2020,df.iloc[i]['borrow_m'],df.iloc[i]['borrow_d']).weekday()<5)

    if i % 1000000 == 0 :
        print('{} per {} is done'.format(i, df.shape[0]))
```

```python
df['isweek'] = df_isweek
```

```python
df.to_csv('df_week20.csv')
```
