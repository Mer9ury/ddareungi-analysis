```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('공공자전거 대여소 정보(21.01.31 기준).csv', encoding='cp949')
df=pd.DataFrame(df)
df
```

```python
df.info()
```

```python
df=df.dropna(subset=['위도', '경도'])
df
```

```python
stn=df.copy()[['대여소 번호', '보관소(대여소)명', '자치구', '위도', '경도']]
stn
```

# 2019 전 기간

```python
brw19=pd.read_csv('brw19.csv')
rtn19=pd.read_csv('rtn19.csv')
brw19=pd.DataFrame(brw19)
rtn19=pd.DataFrame(rtn19)
INNER_JOIN_1=pd.merge(brw19,rtn19, left_on='station', right_on='station', how='inner')
df19=INNER_JOIN_1.copy()[['station','N_out','N_in']]
df19
```

```python
INNER_JOIN_2=pd.merge(stn,df19, left_on='대여소 번호', right_on='station', how='inner')
df19_1=INNER_JOIN_2.copy()[['station','보관소(대여소)명','자치구', '위도', '경도','N_out','N_in']]
df19_1['Count']=df19_1['N_in']+df19_1['N_out']
df19_1['Flow']=(df19_1['N_in']-df19_1['N_out'])/(df19_1['N_in']+df19_1['N_out'])
df19_1
```

```python
deg19=pd.read_csv('degree19.csv')
pg19=pd.read_csv('pagerank19.csv')
deg19=pd.DataFrame(deg19)
pg19=pd.DataFrame(pg19)
INNER_JOIN_3=pd.merge(df19_1,deg19, left_on='station', right_on='station', how='inner')
INNER_JOIN_4=pd.merge(INNER_JOIN_3,pg19, left_on='station', right_on='station', how='inner')
INNER_JOIN_4
df19_2=INNER_JOIN_4.copy()[['station','보관소(대여소)명','자치구', '위도', '경도','N_out','N_in', 'Count','Flow', 'degree_centrality', 'pagerank']]
df19_2=df19_2.sort_values(by='Count', ascending=False)
df19_2
```

```python
df19_2.to_csv('2019년 전 기간 대여소 정리.csv')
```

```python
df19_2.head(15)
```
