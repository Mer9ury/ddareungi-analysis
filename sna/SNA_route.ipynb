```python
import networkx as nx
import matplotlib.pyplot as plt
import csv
import pandas as pd
from collections import Counter
import numpy as np
```

```python
a=pd.read_csv('dataset_2019.csv')
b=pd.read_csv('dataset_2020.csv')
data19=a[['index','borrow_station','return_station','count']]
data20=b[['index','borrow_station','return_station','count']]
data19=data19.sort_values(by='count' ,ascending=False)
data20=data20.sort_values(by='count' ,ascending=False)
```

```python
c=pd.read_csv('공공자전거 대여소 정보(21.01.31 기준).csv', encoding='cp949')
stat=c[['대여소 번호', '보관소(대여소)명','자치구']]
stat
```

```python
d=pd.merge(data19,stat, left_on='borrow_station', right_on='대여소 번호', how='inner')
d=d[['index','borrow_station','return_station','보관소(대여소)명','자치구','count']]
e=pd.merge(d,stat, left_on='return_station', right_on='대여소 번호', how='inner')
df19=e[['index','borrow_station','보관소(대여소)명_x','자치구_x','return_station','보관소(대여소)명_y','자치구_y','count']]
df19.drop_duplicates(['index'])
df19=df19.sort_values(by='count' ,ascending=False)
```

## 2019년 최다 이용 경로

```python
d=pd.merge(data20,stat, left_on='borrow_station', right_on='대여소 번호', how='inner')
d=d[['index','borrow_station','return_station','보관소(대여소)명','자치구','count']]
e=pd.merge(d,stat, left_on='return_station', right_on='대여소 번호', how='inner')
df20=e[['index','borrow_station','보관소(대여소)명_x','자치구_x','return_station','보관소(대여소)명_y','자치구_y','count']]
df20.drop_duplicates(['index'])
df20=df20.sort_values(by='count' ,ascending=False)
```

## 2020년 최다 이용 경로

```python
df20.head(50)
```

```python
df20['index']="("+df20['borrow_station'].astype(str)+", "+df20['return_station'].astype(str)+")"
df20
```

```python
f=pd.merge(df19,df20, left_on='index', right_on='index', how='inner')
df=f[['index','borrow_station_x','보관소(대여소)명_x_x','자치구_x_x','return_station_x','보관소(대여소)명_y_x','자치구_y_x','count_x','count_y']]
df=df.drop_duplicates(['index'])
df['change']=df['count_y']-df['count_x']
```

## 코로나 전후 이용량 증가 경로

```python
df=df.sort_values(by='change' ,ascending=False)
df.head(10)
```
