```python
# 대여소 정보
import pandas as pd
stations = pd.read_csv('./data/공공자전거 대여소 정보(21.01.31 기준).csv', encoding='cp949')
stations = stations[['대여소 번호', '보관소(대여소)명', '위도', '경도']]
stations = stations.dropna()
stations
```

```python
# 대여 이력 정보 2019
import pandas as pd
import os
import numpy as np0
rental2019s = []
print(os.listdir('./data/대여이력2019'))
for file in os.listdir('./data/대여이력2019'):
  print(file)
  try:
    df_2019 = pd.read_csv('./data/대여이력2019/' + file, encoding='cp949')
    df_2019 = df_2019[['대여 대여소번호', '반납대여소번호']]
    df_2019 = df_2019.dropna()
    rental2019s.append(df_2019)
  except Exception as e:
    print(e)
rental2019 = pd.concat(rental2019s)
rental2019_counts = rental2019['대여 대여소번호'].value_counts()
print(rental2019_counts)
for station_num, count in enumerate(rental2019_counts):
  print(station_num, count)
return2019_counts = pd.DataFrame(data=rental2019['반납대여소번호'].value_counts(), columns=['대여소 번호', '반납 대여소번호'])
print(rental2019_counts)
stations = pd.merge(stations, rental2019_counts, how='left', left_on='대여소 번호', right_on='대여소 번호')
stations = pd.merge(stations, return2019_counts, how='left', left_on='대여소 번호', right_on='대여소 번호')
stations
```
