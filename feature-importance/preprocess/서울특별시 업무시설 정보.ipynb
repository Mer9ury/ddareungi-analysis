```python
filename_workplace = '서울특별시 업무시설 정보.csv'
import pandas as pd
workplaces = pd.read_csv(dir+filename_workplace)
workplaces = workplaces[['사업장명', '소재지전체주소', '도로명전체주소', '위치정보(X)', '위치정보(Y)']]
workplaces = workplaces.dropna()
workplaces
```

```python
import requests
from urllib.parse import urlparse
def addr2coord(address):
    url = 'https://dapi.kakao.com/v2/local/search/address.json?query='+address
    headers = {"Authorization": "KakaoAK 61c26d87fa0a7e09587f12e4056fd0da"}
    result = requests.get(urlparse(url).geturl(), headers = headers)
    json_obj = result.json()
    # for document in json_obj['documents'][0]:
    #   val = [document['address_name'], document['y'], document['x']]
    try:
      return [json_obj['documents'][0]['y'],json_obj['documents'][0]['x']]
    except :
      return [None, None]
  
```

```python
dir = 'drive/MyDrive/데이터마이닝/project/Regression/data/'
file_ele = '서울특별시 초등학교 기본정보.csv'
elementary = pd.read_csv(dir+file_ele, encoding='cp949')
elementary = elementary[['학교명','도로명주소']]
# elementary

lats = []
longs = []
count=0
for school in elementary.iloc:
    print(count)
    coord = addr2coord(school['도로명주소'])
    lats.append(coord[0])
    longs.append(coord[1])
    count+=1
elementary['경도'] = lats
elementary['위도'] = longs
elementary
```

```python
file_middle = '서울특별시 중학교 기본정보.csv'
middle = pd.read_csv(dir+file_middle, encoding='cp949')
middle = middle[['학교명','도로명주소']]

lats = []
longs = []
count=0
for school in middle.iloc:
    print(count)
    coord = addr2coord(school['도로명주소'])
    lats.append(coord[0])
    longs.append(coord[1])
    count+=1
middle['경도'] = lats
middle['위도'] = longs
middle
```

```python
flie_high = '서울특별시 고등학교 기본정보.csv'
high = pd.read_csv(dir+flie_high, encoding='cp949')
high = high[['학교명','도로명주소']]

lats = []
longs = []
count=0
high = high.dropna()
for school in high.iloc:
    print(count)
    coord = addr2coord(school['도로명주소'])
    lats.append(coord[0])
    longs.append(coord[1])
    count+=1
high['경도'] = lats
high['위도'] = longs
high
```

```python
flie_univ = '서울시 대학 및 전문대학 DB 정보.csv'
univ = pd.read_csv(dir+flie_univ, encoding='cp949')
univ
univ = univ[['학교명','주소']]

lats = []
longs = []
count=0
for school in univ.iloc:
    print(count)
    coord = addr2coord(school['주소'])
    lats.append(coord[0])
    longs.append(coord[1])
    count+=1
univ['경도'] = lats
univ['위도'] = longs
univ = univ.dropna()
```
