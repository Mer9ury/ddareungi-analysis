```python
# 대여소 정보
import pandas as pd
stations = pd.read_csv('./data/공공자전거 대여소 정보(21.01.31 기준).csv', encoding='cp949')
stations = stations[['대여소 번호', '보관소(대여소)명', '위도', '경도']]
stations = stations.dropna()
stations
```

```python
# 대여소 정보에 TM좌표 추가
%pip install pyproj
from pyproj import Proj, transform
import numpy as np
proj_WGS84 = Proj(init='epsg:4326')
proj_UTMK = Proj(init='epsg:2097')
xs, ys = transform(proj_WGS84, proj_UTMK, np.array(stations['경도']), np.array(stations['위도']))
stations['X좌표'] = xs
stations['Y좌표'] = ys
stations
```

```python
# 대여소 정보에 주변 거주인구 수 추가
import pandas as pd
population = pd.read_csv('./data/대여소별 주변 인구.csv', encoding='euc-kr')
population = population.drop(columns=['Unnamed: 0', '보관소(대여소)명', '위도', '경도', 'X좌표', 'Y좌표', '행정동', '기간', '자치구', '동', '구분'])
stations = pd.merge(left=stations, right=population, how='left', left_on='대여소 번호', right_on='대여소 번호')
stations
```

```python
# 대여소 정보에 주거지역 추가
%pip install geopandas
import pandas as pd
import geopandas as gpd
from shapely.geometry import Point, Polygon, LineString
from math import sqrt
residents = gpd.GeoDataFrame.from_file('./data/UPIS_C_UQ111(주거지역).shp', encoding='cp949')
resident_centers = []
for resident in residents.iloc:
    center = resident['geometry'].centroid
    resident_centers.append([center.x, center.y])
resident_centers = pd.DataFrame(resident_centers)
resident_500s = []
resident_nearests = []
for station in stations.iloc:
    centers_filtered = resident_centers[(abs(resident_centers[0] - station['X좌표']) < 1000) & (abs(resident_centers[1] - station['Y좌표']) < 1000)]
    under_500 = 0
    min_dist = 1500
    for center in centers_filtered.iloc:
        dist = sqrt((center[0] - station['X좌표']) ** 2 +(center[1] - station['Y좌표']) ** 2)
        if dist < 500:
            under_500 += 1
        if dist < min_dist:
            min_dist = dist
    resident_500s.append(under_500)
    resident_nearests.append(min_dist)
stations['주거지역500'] = resident_500s
stations['주거지역최단'] = resident_nearests
stations
```

```python
# 대여소 정보에 업무시설 추가
import pandas as pd
from math import sqrt
workplaces = pd.read_csv('./data/서울특별시 업무시설 정보.csv', encoding='utf8')
workplaces = workplaces[['위치정보(X)', '위치정보(Y)']]
workplaces = workplaces.dropna()
workplace_500s = []
workplace_nearests = []
for station in stations.iloc:
    workplaces_filtered = workplaces[(abs(workplaces['위치정보(X)'] - station['X좌표']) < 1000) & (abs(workplaces['위치정보(Y)'] - station['Y좌표']) < 1000)]
    under_500 = 0
    min_dist = 1500
    for workplace in workplaces_filtered.iloc:
        dist = sqrt((workplace['위치정보(X)'] - station['X좌표']) ** 2 +(workplace['위치정보(Y)'] - station['Y좌표']) ** 2)
        if dist < 500:
            under_500 += 1
        if dist < min_dist:
            min_dist = dist
    workplace_500s.append(under_500)
    workplace_nearests.append(min_dist)
stations['업무시설500'] = workplace_500s
stations['업무시설최단'] = workplace_nearests
stations
```

```python
# 대여소 정보에 지하철역 추가
%pip install haversine
import pandas as pd
from haversine import haversine
subways = pd.read_csv('./data/station_coordinate.csv', encoding='utf8')
subways = subways[['lat', 'lng']]
subway_500s = []
subway_nearests = []
for station in stations.iloc:
    subways_filtered = subways[(abs(subways['lng'] - station['경도']) < 0.01) & (abs(subways['lat'] - station['위도']) < 0.01)]
    under_500 = 0
    min_dist = 1.5
    for subway in subways_filtered.iloc:
        dist = haversine((station['경도'], station['위도']), (subway['lng'], subway['lat']))
        if dist < 0.5:
            under_500 += 1
        if dist < min_dist:
            min_dist = dist
    subway_500s.append(under_500)
    subway_nearests.append(min_dist)
stations['지하철500'] = subway_500s
stations['지하철최단'] = subway_nearests
stations
```

```python
# 대여소 정보에 버스정류장 추가
import pandas as pd
from haversine import haversine
import os
buses = pd.read_csv('./data/서울시버스정류소좌표데이터(2021.01.14.).csv', encoding='cp949')
buses = buses[['ARS-ID', 'X좌표', 'Y좌표']]
buses['ARS-ID-str'] = pd.Series(buses['ARS-ID'], dtype=str)
for idx, id in enumerate(buses['ARS-ID']):
  if id < 10000:
    buses['ARS-ID-str'][idx] = '{:05d}'.format(id)
bus_usages = []
for file in os.listdir('./data/버스이용량2019'):
  bus_usage = pd.read_csv('./data/버스이용량2019/' + file, encoding='cp949')
  bus_usage['이용량2019'] = bus_usage['승차총승객수'] + bus_usage['하차총승객수']
  bus_usage = bus_usage[['버스정류장ARS번호', '이용량2019']]
  bus_usage = bus_usage.groupby(['버스정류장ARS번호'], as_index=False).sum()
  bus_usages.append(bus_usage)
bus_usages = pd.concat(objs=bus_usages)
bus_usages = bus_usages.groupby(['버스정류장ARS번호'], as_index=False).sum()
buses = pd.merge(buses, bus_usages, how='left', left_on='ARS-ID-str', right_on='버스정류장ARS번호')
bus_usages = []
for file in os.listdir('./data/버스이용량2020'):
  bus_usage = pd.read_csv('./data/버스이용량2020/' + file, encoding='cp949')
  try:
    bus_usage['이용량2020'] = bus_usage['승차총승객수'] + bus_usage['하차총승객수']
  except KeyError:
    bus_usage['이용량2020'] = bus_usage['ride_pasgr_num'] + bus_usage['alight_pasgr_num']
  try:
    bus_usage = bus_usage[['버스정류장ARS번호', '이용량2020']]
  except KeyError:
    bus_usage = bus_usage[['bsst_ars_no', '이용량2020']].rename(columns={'bsst_ars_no': '버스정류장ARS번호'})
  bus_usage = bus_usage.groupby(['버스정류장ARS번호'], as_index=False).sum()
  bus_usages.append(bus_usage)
bus_usages = pd.concat(objs=bus_usages)
bus_usages = bus_usages.groupby(['버스정류장ARS번호'], as_index=False).sum()
buses = pd.merge(buses, bus_usages, how='left', left_on='ARS-ID-str', right_on='버스정류장ARS번호')
buses = buses.dropna()
bus_500s = []
bus_nearests = []
bus_usages_2019 = []
bus_usages_2020 = []
for station in stations.iloc:
    buses_filtered = buses[(abs(buses['X좌표'] - station['경도']) < 0.01) & (abs(buses['Y좌표'] - station['위도']) < 0.01)]
    under_500 = 0
    min_dist = 1.5
    bus_usage_2019 = 0
    bus_usage_2020 = 0
    for bus in buses_filtered.iloc:
        dist = haversine((station['경도'], station['위도']), (bus['X좌표'], bus['Y좌표']))
        if dist < 0.5:
            under_500 += 1
            bus_usage_2019 += bus['이용량2019']
            bus_usage_2020 += bus['이용량2020']
        if dist < min_dist:
            min_dist = dist
    bus_500s.append(under_500)
    bus_nearests.append(min_dist)
    bus_usages_2019.append(bus_usage_2019)
    bus_usages_2020.append(bus_usage_2020)
stations['버스500'] = bus_500s
stations['버스최단'] = bus_nearests
stations['버스이용량2019'] = bus_usages_2019
stations['버스이용량2020'] = bus_usages_2020
stations
```
